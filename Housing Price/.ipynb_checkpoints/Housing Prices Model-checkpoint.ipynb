### 3. Prepare Data


```python
##3.a Data Cleaning
data = data.drop(['id', 'date'], axis=1)
#  Lets drop the date and id columns as it does not contribute much to the final price of the house
```

```python
has_yr_renovated = (data['yr_renovated'] != 0).sum()
does_not_have_yr_renovated = (data['yr_renovated'] == 0).sum()
print("Only {}% of the entire dataset has meaningful information about the year renovated".format(str(round(has_yr_renovated/does_not_have_yr_renovated*100,2))))
```

```python
##3.b Feature Selection 
X = data.drop('price', axis=1)
y = data['price']

# Use RandomForestRegressor to get feature importance
model = RandomForestRegressor()
model.fit(X, y)

# Display feature importance
feature_importance = pd.Series(model.feature_importances_, index=X.columns)
feature_importance.sort_values(ascending=False, inplace=True)
print(feature_importance)
```

### Notes 
Taking a closer look at the independent variables we can see that 
1. ***'sqft_living'*** is the sum of ***'sqft_above'*** and ***'sqft_basement'***, So including all three variables in a model might introduce multicollinearity. 
2. ***'sqft_lot'*** and ***'sqft_lot15'*** might be correlated, as the size of the land might have some spatial correlation with the sizes of nearby land lots. 
3. ***'sqft_living15'*** and ***'sqft_lot15'*** might be correlated, as the size of the land might have some spatial correlation with the sizes of nearby land lots. 
4. We know that ***Zipcode*** could be useful as an independent variable but it cannot be used as an integer as is. It needs to be converted into something more meaningful. Realistically I would augment the zipcode with latitude and longtitude to create clusters, but we are going to create a ***Zipcode_index*** variable and drop ***Zipcode*** 


### Decisions 

-The following variables are being dropped from the models
1. ***'sqft_above'*** 
2. ***'sqft_basement'*** 
3. ***'sqft_lot'*** 
4. ***'sqft_lot15'*** 
5. ***'Zipcode'*** - (after converting it into zipcode_index) 
6. ***'yr_built'*** - (low correlation according to the correlation matrix)
7. ***'yr_renovated'*** - (low data availability)

```python
##3.c Data Transforms 
```

```python
# we are going to transform zipcodes into zipcode index. 
data['zipcode'].count()
# we have 21613 datapoints 
```

```python
len(pd.unique(data['zipcode']))
# we have 70 uniques zipcodes
```

```python
zipcodes = list(pd.unique(data['zipcode']))
zipcode_to_index = {zipcode: index for index, zipcode in enumerate(set(zipcodes))}
data['zipcode_index'] = data['zipcode'].map(zipcode_to_index)
```

```python
data = data.drop(['sqft_above', 'sqft_basement','sqft_lot','sqft_lot15','yr_built','yr_renovated','zipcode'], axis=1)
```

```python
corr_matrix = data.corr(numeric_only=True)
plt.figure(figsize=(16, 15))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.show()
```

```python
X = data.drop('price', axis=1)
y = data['price']

# Use RandomForestRegressor to get feature importance
model = RandomForestRegressor()
model.fit(X, y)

# Display feature importance
feature_importance = pd.Series(model.feature_importances_, index=X.columns)
feature_importance.sort_values(ascending=False, inplace=True)
print(feature_importance)
```

We definitely improved the feature importance of the zipcode by converting it and getting rid of the other variables

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

#### Scaling is the transforming of numerical features to a standard range. The purpose of scaling is to ensure that all features contribute equally to the model

### 4. Model Building and Evaluate Algorithms 

```python
#4.a  Split-out validation dataset 
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
#4.b  Test options and evaluation metric
scoring = 'neg_mean_squared_error'
```

#### The mean of the negative mean squared error is calculated, and a higher value indicates better model performance, just as a convention to align with the "greater is better" expectation 

```python
#4.c Spot Check Algorithms
models = [
    ('RandomForest', RandomForestRegressor()),
    ('GradientBoosting', GradientBoostingRegressor()),
    ('LinearRegression', LinearRegression()),
    ('SVR', SVR()),
    ('XGBoost', XGBRegressor())
]


#4.d Compare Algorithms 
results = []
names = []
for name, model in models:
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    mse = mean_squared_error(y_test, predictions)
    results.append(mse)
    names.append(name)
    print(f'{name}: MSE - {mse}')
```

### 5. Improve Accuracy 

```python
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_search = GridSearchCV(RandomForestRegressor(), param_grid, scoring=scoring)
grid_result = grid_search.fit(X_train, y_train)

best_params = grid_result.best_params_
best_model = grid_result.best_estimator_
```

```python
#Ensembles 
ensemble_models = [('RandomForest', RandomForestRegressor()), ('XGBoost', XGBRegressor())]

ensemble = VotingRegressor(estimators=ensemble_models)
ensemble.fit(X_train, y_train)
ensemble_predictions = ensemble.predict(X_test)
ensemble_mse = mean_squared_error(y_test, ensemble_predictions)
print(f'Ensemble MSE: {ensemble_mse}')
```

### 6. Finalize Model 

```python
#5.a Predictions on validation dataset 
final_predictions = best_model.predict(X_test)
```

```python
#5.b Create standalone model on entire training dataset
best_model.fit(X_scaled, y)
```

```python
#5.c Save model for later use 
joblib.dump(best_model, 'final_model.joblib')
```
